```python
ipd.Audio(audio_files[0])
ipd.Audio(audio_files[1])
ipd.Audio(audio_files[2])
```

```python
y0, sr = librosa.load(audio_files[0])
y1, _ =  librosa.load(audio_files[1])
y2, _ =  librosa.load(audio_files[2])
```

```python
pd.Series(y0).plot(figsize=(10,5), lw=1, title="roar audio example")
plt.show()
```

```python
pd.Series(y1).plot(figsize=(10,5), lw=1, title="iroar audio example")
plt.show()
```

```python
pd.Series(y2).plot(figsize=(10,5), lw=1, title="grunt audio example")
plt.show()
```

```python
y0_trimmed, _ = librosa.effects.trim(y0)

pd.Series(y0_trimmed[0:2000]).plot(figsize=(10,5), lw=1, title="roar audio example")
plt.show()
```

```python
y1_trimmed, _ = librosa.effects.trim(y1)

pd.Series(y1_trimmed[0:2000]).plot(figsize=(10,5), lw=1, title="iroar audio example")
plt.show()
```

```python
y2_trimmed, _ = librosa.effects.trim(y2)

pd.Series(y2_trimmed[0:2000]).plot(figsize=(10,5), lw=1, title="grunt audio example")
plt.show()
```
